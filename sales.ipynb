```python
import pandas as pd

file_path = r'D:\1819intern\task2/Superstore1.csv' 

# FIX: Add the 'encoding' parameter, usually 'latin-1' or 'cp1252'
try:
    df = pd.read_csv(file_path, encoding='latin-1')
    print("File read successfully with latin-1 encoding.")
except UnicodeDecodeError:
    # If latin-1 fails, try 'cp1252' (a very common Windows encoding)
    df = pd.read_csv(file_path, encoding='cp1252')
    print("File read successfully with cp1252 encoding.")

print(df.info())
```

```python
df.dropna(inplace=True)
```

```python
# Convert to datetime objects
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='mixed')

print(df[['Order Date', 'Ship Date']].info()) # Verify the Dtype is now datetime64
```

```python
# Check for truly identical rows (structural errors)
print(f"Total fully duplicate rows: {df.duplicated().sum()}")

# Remove structural duplicates if any are found
df.drop_duplicates(inplace=True)
```

```python
# Calculate shipping time in days
df['Shipping_Days'] = (df['Ship Date'] - df['Order Date']).dt.days

# Check the new feature's summary
print(df['Shipping_Days'].describe())
```

```python
# Convert low-cardinality columns
categorical_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']
for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
# 1. Check for True/False (True means missing)
# This returns a DataFrame of booleans
df.isnull().head()

# 2. Total count of missing values for EACH column
# This is the most common and useful command for a quick assessment
print("Missing values per column:")
print(df.isnull().sum())

# 3. Total count of ALL missing values in the entire DataFrame
print(f"\nTotal missing values in the entire DataFrame: {df.isnull().sum().sum()}")

# 4. Show the percentage of missing values per column
print("\nPercentage of missing values per column:")
print((df.isnull().sum() / len(df)) * 100)
```

```python
# Convert low-cardinality columns for efficiency
categorical_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']
for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
df = pd.get_dummies(df, columns=['Segment', 'Ship Mode'], drop_first=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

df['Category_Encoded'] = le.fit_transform(df['Category'])
```

```python
import numpy as np

df['Sales_Log'] = np.log1p(df['Sales'])
```

```python
df['Order_Day_of_Week'] = df['Order Date'].dt.dayofweek
df['Order_Month'] = df['Order Date'].dt.month
df['Sales_by_Quantity'] = df['Sales'] * df['Quantity']
```

```python
# Final check before export
print(df.info())

import pandas as pd
import os # Import the built-in operating system module

# Define the file path
output_file_path = 'data/processed/superstore_final_for_BI.csv'
# Extract the directory path from the full file path
output_dir = os.path.dirname(output_file_path)

# --- FIX START ---
# Check if the directory exists, and if not, create it
if not os.path.exists(output_dir):
    # This creates all intermediate folders ('data' and 'processed')
    os.makedirs(output_dir) 
    print(f"Created directory: {output_dir}")
# --- FIX END ---

# Now the export will work
df.to_csv(output_file_path, index=False)

print("\n✅ Successfully exported final cleaned data for visualization tools.")
```

```python
# Final Python Action: Export the Prepared Data

# Define the file path for the output CSV
output_file_path = 'data/processed/superstore_final_for_BI.csv'

# Export the DataFrame
# index=False ensures the DataFrame index (0 to 9993) is not saved as an extra column.
df.to_csv(output_file_path, index=False)

print(f"\n✅ Data preparation complete!")
print(f"Data successfully exported to: {output_file_path}")
print("You can now import this CSV file into Tableau or Power BI.")
```
